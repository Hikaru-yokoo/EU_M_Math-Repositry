```python
#y = sin x とy = cos xを重ねる(-10 < x < 10)

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def function_sin(x):
    return np.sin(x)

def function_cos(x):
    return np.cos(x)

x = np.linspace(-10,10,100)
plt.plot(x, function_sin(x))
plt.plot(x, function_cos(x))
plt.grid(True)
```
